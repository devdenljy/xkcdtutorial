```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.linspace(0.1,10.00)
```

```python
with plt.xkcd():
    
  left, width = .25, .5
  bottom, height = .25, .5
  right = left + width
  top = bottom + height

  fig = plt.figure()

  plt.plot(np.linspace(0.1, 10.00), np.linspace(0.1, 10.00))
  plt.text(3.5*(left+right), 5.0*(bottom+top),'$y=ax+c$')
```

```python
with plt.xkcd():
    
  left, width = .25, .5
  bottom, height = .25, .5
  right = left + width
  top = bottom + height

  fig, ax = plt.subplots()
    
  ax.plot(np.linspace(0.1, 10.00), 2**np.linspace(0.1, 10.00), '-')
  plt.text(4.5*(left+right), 500*(bottom+top),'$y=k^{x}$')
```

```python
np.geomspace(0.1, 10.00)
```

```python
np.logspace(1.0, 10.00)
```

```python
with plt.xkcd():
  fig, ax = plt.subplots()

  ax.plot(np.logspace(1.0, 10.00), np.geomspace(0.1, 10.00), '-')
  plt.text(0.5, 5.0, '$y=f(k)$')
```

```python
np.logspace(1.0, 10.00)/np.geomspace(0.1, 10.00)
```

```python
with plt.xkcd(): 
  fig, ax = plt.subplots()

  ax.plot(np.logspace(1.0, 10.00), np.logspace(1.0, 10.00)/np.geomspace(0.1, 10.00), '-')
```
